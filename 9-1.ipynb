```python
import pandas as pd
```

```python
dia = pd.read_csv("diamonds.csv")
dia.head(2)
```

```python
pd.crosstab(dia["cut"],dia["clarity"])
```

```python
pd.crosstab(dia["cut"],dia["clarity"],normalize=True)
```

```python
pd.crosstab(dia["cut"],dia["clarity"],normalize=True).round(2)
```

```python
pd.crosstab(dia["cut"],dia["clarity"],normalize=0)
```

```python
pd.crosstab(dia["cut"],dia["clarity"],normalize=1) #wide form
```

```python
#pd.crosstab(dia["cut"],dia["clarity"],values=dia["price"],
#           aggfunc=pd.Series.mean)
dia.groupby(["cut","clarity"])["price"].mean().reset_index() #long form
```

```python
#sort values
```

```python
dia_agg = dia.groupby(["cut","clarity"])["price"].mean().reset_index()
dia_agg.head()
```

```python
dia_agg.sort_values("price").head() 
```

```python
dia_agg.sort_values("price",ascending = False).head() 
```

```python
dia_agg.sort_values(["cut","clarity"],ascending =[True,False]).head() 
```
